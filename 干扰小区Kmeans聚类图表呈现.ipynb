
#### kmeans步骤：

（1）随机找 k 个点作为质心（种子）；

（2）计算其他点到这 k 个种子的距离，选择最近的那个作为该点的类别；

（3）更新各类的质心，迭代到质心的不变为止。

Q:如何选择 k 值？

A: 根据 k 取不同的值时，模型性能曲线图。 横坐标是聚类数目k，纵坐标是各点到距离中心的距离和。在拐点处的K（距离和下降速率变慢）作为选择的值。
